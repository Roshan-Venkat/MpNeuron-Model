#Binarization of data
Mp neuron can only take 0's and 1's as input 

```
X_binary_train = Xtrain.apply(pd.cut, bins = 2, labels = [1,0]) #[1,0] beacuse of the trend, where bening tumours have high values and malignant have low values as per data
```

```
X_binary_test = Xtest.apply(pd.cut, bins = 2, labels = [1,0]) #[1,0] beacuse of the trend, where bening tumours have high values and malignant have low values as per data
```

```
plt.plot(X_binary_test.T,'*')
plt.xticks(rotation = 'vertical')
plt.show()
```

```
X_binary_test = X_binary_test.values
X_binary_train = X_binary_train.values
```

```
print(X_binary_train.shape, X_binary_test.shape)
```

#MP neuron Model

```
max_acc = 0;
max_acc_b = 0;
for b in range(X_binary_train.shape[1] + 1):
  Ypredicted = []
  accurate = 0
  for x , y in zip(X_binary_train,Ytrain):
    Ypredicted.append(np.sum(x) >= b)
    accurate += (y == (np.sum(x) >= b))
  print(b , accurate / X_binary_train.shape[0])
  if max_acc < (accurate /X_binary_train.shape[0]):
    max_acc = (accurate /X_binary_train.shape[0])
    max_acc_b = b
print("Maximum accuracy obtained : " ,max_acc)
print("B value : " , max_acc_b)   
```

```
from sklearn.metrics import accuracy_score
```

```
Ypredicted_test = []

for x in X_binary_test:
  Ypredicted_test.append(np.sum(x) >= max_acc_b)

print(accuracy_score(Ypredicted_test,Ytest))
```
